```python
import numpy as np
import rasterio
from rasterio.transform import from_origin
```

```python
def create_hrelease_raster(length, width, thickness, center_x, center_y, filename):
    
    # Cell size
    dx = 40 
    dy = 40 
    
    cols = int(length / dx)
    rows = int(width / dy)
    
    Z_release = np.full((rows, cols), thickness, dtype=np.float32)
    
    top_left_x = center_x - (length / 2)

    top_left_y = center_y + (width / 2)
    
    transform = from_origin(top_left_x, top_left_y, dx, dy) 
    
    with rasterio.open(
        filename, 'w', driver='GTiff',
        height=Z_release.shape[0], width=Z_release.shape[1],
        count=1, dtype=Z_release.dtype,
        crs='EPSG:32643',
        transform=transform
    ) as dst:
        dst.write(Z_release, 1)
        
    print(f"Generated {filename}: {length}m x {width}m, Thickness: {thickness}m")
    
Cx = 681.4 
Cy = 547.9

# Run A: Baseline
create_hrelease_raster(length=200, width=60, thickness=20, center_x=Cx, center_y=Cy, filename="Landslide_40m.tif")
```
